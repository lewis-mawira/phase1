# 1. Defining the Question
 
   ### a) Specifying the Data Analytic Question
   
The movies industry is characterized by diffrent genres of movies. They include animations, action movies, comedies etc. When producing a movie it is important to decide on what genre to get into. This can be done by determining the best perfoming movies in the market. There are different aspects that can be used to determine how different genres are perfoming example What are the top-perfoming movie genres in terms of box office revenue, and how can this information be used to guide Microsoft's new movie studio in selecting the most promising genre for their film production?
  
  ### b) Defining the Metrics for Success
The primary objective is to identify movie genres that have achieved remarkable commercial success based on box office revenue. Analyzing and comparing revenue figures across various genres will enable us to determine the genres with a higher likelihood of achieving commercial success. Additionally, metrics such as the involvement of popular actors in highly rated movies, as well as the directors and writers associated with the best-rated films, can be considered to evaluate success.

We will consider the following key aspects:

```python
new_bom_gross.info()
```

```python
new_bom_gross.head()
```

There are missing values in the two rows domestic_gross and forign_gross of type floats. We can convert them into integers so that we can calculate its measures of central tendency and replace the missing values appropriately

There are columns containing non-finite values so I first had to replace them with the mode to allow the conversion, there were also commas and whitespaces so i had to handle that too.

```python
new_bom_gross.info()
```

```python
# Fill missing values with the mode in 'domestic_gross' column
domestic_mode = new_bom_gross['domestic_gross'].mode().values[0]
new_bom_gross['domestic_gross'].fillna(domestic_mode, inplace=True)

# Fill missing values with the mode in 'foreign_gross' column
foreign_mode = new_bom_gross['foreign_gross'].mode().values[0]
new_bom_gross['foreign_gross'].fillna(foreign_mode, inplace=True)

# Convert 'domestic_gross' column to string
new_bom_gross['domestic_gross'] = new_bom_gross['domestic_gross'].astype(str)

# Convert 'foreign_gross' column to string
new_bom_gross['foreign_gross'] = new_bom_gross['foreign_gross'].astype(str)

# Remove commas from 'domestic_gross' column values and convert to float
new_bom_gross['domestic_gross'] = new_bom_gross['domestic_gross'].str.replace(',', '').astype(float)

# Remove commas from 'foreign_gross' column values and convert to float
new_bom_gross['foreign_gross'] = new_bom_gross['foreign_gross'].str.replace(',', '').astype(float)
```

```python
new_bom_gross.info()
```

All our columns now are populated and we have converted the domestic_gross and foreign_gross columns into strings for plotting purposes

## Plot between Genres and Domestic Gross Income

## Plot between Genres and Worldwide Gross Income.

```python
# Group the data by genre and calculate the sum of worldwide gross
genre_worldwide_gross = merged_budget_Tmdb_df.groupby('genre')['worldwide_gross'].sum()

# Sort the genres based on the original order
genre_worldwide_gross = genre_worldwide_gross.reindex(merged_budget_Tmdb_df['genre'].unique())

# Plot the bar chart
plt.figure(figsize=(8, 10))
plt.bar(genre_worldwide_gross.index, genre_worldwide_gross.values)

# Set the title and labels
plt.title('Worldwide Gross by Genre')
plt.xlabel('Genre')
plt.ylabel('Worldwide Gross')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Display the plot
plt.show()
```

## Creating a Subplot side by side for better visualizations and interpretetion

```python
# Group the data by genre and calculate the sum of domestic gross
genre_domestic_gross = merged_budget_Tmdb_df.groupby('genre')['domestic_gross'].sum()

# Sort the genres based on the original order
genre_domestic_gross = genre_domestic_gross.reindex(merged_budget_Tmdb_df['genre'].unique())

# Group the data by genre and calculate the sum of worldwide gross
genre_worldwide_gross = merged_budget_Tmdb_df.groupby('genre')['worldwide_gross'].sum()

# Sort the genres based on the original order
genre_worldwide_gross = genre_worldwide_gross.reindex(merged_budget_Tmdb_df['genre'].unique())

# Determine the maximum value for the y-axis limit
max_gross = max(genre_domestic_gross.max(), genre_worldwide_gross.max())

# Create a figure with two subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 8))

# Plot the bar chart for domestic gross in the first subplot
axs[0].bar(genre_domestic_gross.index, genre_domestic_gross.values)
axs[0].set_title('Domestic Gross by Genre')
axs[0].set_xlabel('Genre')
axs[0].set_ylabel('Domestic Gross')
axs[0].tick_params(axis='x', rotation=90)
axs[0].set_ylim(0, max_gross)  # Set the y-axis limit

# Plot the bar chart for worldwide gross in the second subplot
axs[1].bar(genre_worldwide_gross.index, genre_worldwide_gross.values)
axs[1].set_title('Worldwide Gross by Genre')
axs[1].set_xlabel('Genre')
axs[1].set_ylabel('Worldwide Gross')
axs[1].tick_params(axis='x', rotation=90)
axs[1].set_ylim(0, max_gross)  # Set the y-axis limit

# Adjust the spacing between subplots
plt.tight_layout()
# Save the plot with a file name
plt.savefig('comparison.GENRES.png')
# Display the plots
plt.show()

```

Different movie genres exhibit varying levels of performance in terms of gross domestic income and gross worldwide income. This observation suggests that audience reception and market success are influenced by the genre of the movie. Some genres tend to generate higher domestic and worldwide income, indicating a stronger appeal and popularity among audiences. This insight highlights the importance of considering genre when making decisions related to film production, distribution, and targeting specific markets.

## **Checking for correlation between gross domestic income and gross foreign income**

```python
# Extracting domestic gross and foreign gross data
domestic_gross = final_merged_budget['domestic_gross']
foreign_gross = final_merged_budget['foreign_gross']

# Creating a scatter plot
plt.scatter(domestic_gross, foreign_gross, alpha=0.5)

# Adding labels and title
plt.xlabel('Domestic Gross')
plt.ylabel('Foreign Gross')
plt.title('Correlation of Domestic Gross and Foreign Gross')

# Show the plot
plt.show()
```

```python
correlation = np.corrcoef(domestic_gross, foreign_gross)[0, 1]
print(f'The correlation coefficient is: {correlation}')
```

A correlation coefficient of 0.78 suggests a **strong positive** linear relationship between the **domestic gross income and foreign gross income** variables. This means that as the domestic gross increases, the foreign gross also tends to increase, and vice versa. 
From the scatter plot, the values cluster together and move up from left to right, indicating that as the domestic gross increases, the foreign gross tends to increase as well.
